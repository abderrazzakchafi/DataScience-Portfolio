```python
import sqlite3
import pandas as pd

df = pd.read_csv("cleaned_employees_data.csv")
conn = sqlite3.connect("employees.db")
df.to_sql("employees", conn, if_exists="replace", index=False)

print("Database created ")
```

```python
query ="""
SELECT *
FROM employees
WHERE department = "IT"
LIMIT 5;
"""

result = pd.read_sql_query(query, conn)

display(result)
```

```python
query="""
SELECT *
FROM employees
WHERE salary >= 4000
LIMIT 5;
"""

result = pd.read_sql_query(query, conn)
display(result)
```

```python
query = """
SELECT name, salary
FROM employees
WHERE department = "sales"
"""
result = pd.read_sql_query(query, conn)
display(result)
```

```python
query= """
SELECT name, salary
FROM employees
WHERE department = "sales"
ORDER BY salary DESC
LIMIT 5
"""
result = pd.read_sql_query(query, conn)
display(result)
```

```python
query= """
SELECT AVG(salary)
FROM employees
WHERE department = "IT"
"""
result = pd.read_sql_query(query, conn)
display(result)
```
