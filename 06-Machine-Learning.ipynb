```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('cleaned_employees_data.csv')

df.head(3)
```

```python
df = df.drop(['id', 'email', 'name' ], axis=1, errors='ignore')
df.head(3)
```

```python
df.isnull().sum()
```

```python
import numpy as np
df.replace('unknown', np.nan, inplace=True)
df.head(3)
```

```python
df.isnull().sum()
```

```python
print((df.isnull().sum()/ len(df))* 100)
```

```python
df['country'] = df['country'].fillna("others")
df['city'] = df['city'].fillna('others')
df['department'] = df['department'].fillna(df['department'].mode()[0])
df.head(3)
```

```python
df.isnull().sum()
```

```python
df_final = pd.get_dummies(df, drop_first=True)
df_final.head()
```

```python
df['country'] = df['country'].replace(['fr','fra'], 'france')
df_final = pd.get_dummies(df)
df_final.head()
```
