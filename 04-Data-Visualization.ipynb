```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv("cleaned_employees_data.csv")

print("number of employees is: ", df.shape[0])
df.head(10)
```

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df, x="department", hue="department", palette="viridis", legend=False)

plt.title("Number of Employees per Department", fontsize=16, fontweight="bold")
plt.xlabel("Department", fontsize=12)
plt.ylabel("Employee Count", fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=df, x="salary", bins=30, kde=True, color="royalblue")

plt.title("Salary Distribution", fontsize=16, fontweight="bold")
plt.xlabel("Salary Euro", fontsize=12)
plt.ylabel("number of Employees", fontsize=12)

plt.show()
```
