```python
import pandas as pd

df = pd.read_csv("very_messy_dataset_2500_rows.csv")
df.head(10)
```

```python
text_columns = ["city", "name", "country", "department"]

for col in text_columns:
    df[col] = df[col].str.lower().str.strip()

df.head(5)
```

```python
df["salary"] = df["salary"].str.replace("€", "", regex=False)
df["salary"] = pd.to_numeric(df["salary"], errors="coerce")

df["salary"].head(10)
```

```python
df["age"] = pd.to_numeric(df["age"], errors="coerce")
df["age"].head(10)
```

```python
print(f"NaT %: {df['join_date'].isnull().mean() * 100:.2f}%")
```

```python
df["join_date"] = pd.to_datetime(df["join_date"], errors="coerce")
df["join_date"].head(10)
```

```python
df = df.drop(columns=["join_date"])
print(df.columns)
```

```python
df["score"] = pd.to_numeric(df["score"], errors="coerce")
df["score"].head(10)
```

```python
Q1 = df["salary"].quantile(0.25)
Q3 = df["salary"].quantile(0.75)

IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
print(f"the upper limit is: {upper_limit}")

df = df[df["salary"] <= upper_limit]
print(df["salary"].max())

df = df[df["salary"] > 0]

print("befor:", df.shape[0])
df = df[df["salary"] <= upper_limit]
print("after", df.shape[0])
```
